```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
np.random.seed(42)
#Radom data set for PR to PO Operation 

#for 1000 Lines
n = 1000

data = {
    "PO_ID": range(10001, 10001 + n),
    "Supplier_ID": np.random.randint(200, 350, n),
    "Category": np.random.choice(["Direct", "Indirect"], n),
    "Region": np.random.choice(["NA", "EMEA", "APAC"], n),
    "Unit_Price": np.round(np.random.uniform(10, 500, n), 2),
    "Quantity": np.random.randint(1, 1000, n),
    "Lead_Time_Days": np.random.randint(5, 60, n),
    "On_Time_Delivery": np.random.choice([0, 1], n, p=[0.3, 0.7]),
    "Quality_Score": np.round(np.random.uniform(2.5, 5.0, n), 2)
}

df = pd.DataFrame(data)

df.head()
```

```python
#Trying to answer below question
#1. Which suppliers behave similarly?
#2. How do we create supplier personas for sourcing strategy?

#Features Used for Clustering
cluster_features = [
    "Unit_Price",
    "Quantity",
    "Lead_Time_Days",
    "On_Time_Delivery",
    "Quality_Score"
]
```

```python


X_cluster = df[cluster_features]

# Scale data (VERY IMPORTANT for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)
```

```python

#The “elbow” point (usually K=3 or K=4) is optimal
#Beyond this, improvement is marginal
wcss = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df["Supplier_Cluster"] = kmeans.fit_predict(X_scaled)
#Each supplier is now assigned to a cluster (0,1,2)
df.head()
```

```python
cluster_summary = df.groupby("Supplier_Cluster")[cluster_features].mean()
cluster_summary
```
